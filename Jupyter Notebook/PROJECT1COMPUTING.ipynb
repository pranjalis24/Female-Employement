###LINEAR REGRESSION

```r
#Model1
plot(mydataset$PerFemEmploy, mydataset$PerFemEmployers, 
  pch = 16, 
  cex = 1.3, 
  col = "blue", 
  main = "relation between female employes and female employers", 
  xlab = "female employes", 
  ylab = "female employers")

model_one <- lm(mydataset$PerFemEmploy ~ mydataset$PerFemEmployers)
model_one=summary(model_one)
abline(model_one)

print(model_one)
```

This LR Model shows that the coefficients and intercept line as given below y= 24.921 + 14.053x1 And p-value: 3.671e-09

```r
#Model2
plot(mydataset$ContrFamWorkers,mydataset$OwnAccount, 
  pch = 16, 
  cex = 1.3, 
  col = "green", 
  main = "own account and family workers", 
  xlab = "fam workers", 
  ylab = "own account")

model_two <- lm(mydataset$OwnAccount ~ mydataset$ContrFamWorkers)
model_two=summary(model_two)
abline(model_two)

print(model_two)
```

This LR Model shows that the coefficients and intercept line as given below y= 58.29050 -0.65008x1 And p-value < 2.2e-16

```r
#Model3
SM1 = lm(mydataset$Agriculture~mydataset$Wage.Salaried, data = mydataset)
SM1 = summary(SM1)
SM1
```

This LR Model shows that the coefficients and intercept line as given below y= 94.2999 -1.0939x1 And p-value:0.0006788

```r
#Model4
plot(
  mydataset$OwnAccount, 
  mydataset$Agriculture, 
  pch = 16, 
  cex = 1.3, 
  col = "green", 
  main = "own account and agriculture", 
  xlab = "own account", 
  ylab = "agriculture")

model_four <- lm(mydataset$Agriculture ~ mydataset$Agriculture)

abline(model_four)
model_four=summary(model_four)
print(model_four)
```

This LR Model shows that the coefficient is 70.272.

###MULTIPLE LINEAR REGRESSION

```r
plot(mydataset$Agriculture, mydataset$Industry+mydataset$Services, 
  pch = 20, 
  cex = 1.8, 
  col = "red", 
  main = "agriculture ,industry,sevices", 
  xlab = "agriculture", 
  ylab = "industry and services")
MR1 = lm(mydataset$Agriculture ~ mydataset$Industry + mydataset$Services, data = mydataset)
MR1 = summary(MR1)
MR1
abline(MR1)

print(MR1)
```

The Multiple Linear Regression Model here is a relationship between multiple variables- Female employement in three sectors- Agriculture, Industry and Services. This LR Model shows that the coefficients and intercept line as given below y= 1.000 + (-9.996)x1 + (-1.000)x2 And p-value < 2.2e-16

```r
plot(
  mydataset$Year, 
  mydataset$Wage.Salaried+mydataset$Vulnerable+mydataset$Vulnerable, 
  pch = 20, 
  cex = 1.8, 
  col = "red", 
  main = "agriculture ,industry,sevices", 
  xlab = "agriculture", 
  ylab = "industry and services")
MR2 = lm(mydataset$Year ~ mydataset$Wage.Salaried + mydataset$Vulnerable+mydataset$Vulnerable, data = mydataset)
MR2 = summary(MR2)
MR2
abline(MR2)

print(MR2)
```

The Multiple Linear Regression Model here is a relationship between multiple variables- Female employement in three sectors to - Agriculture, Industry and Services. This LR Model shows that the coefficients and intercept line as given below y= 3455.37 + (-14.03)x1 + (-14.65)x2 And p-value: 0.0003324
